```python
df = pd.get_dummies(df, columns=["fuel", "aspiration", "doors", "drive-wheel", "engine-loc", "body", 'engine-type', 'cylinders', 'fuel-sys'])
```

```python
import category_encoders as ce
make_be = ce.BinaryEncoder(cols=['make'])
make = make_be.fit_transform(df['make'])

for i in range(make.shape[1]):
    df[f'make_{i}'] = make[f'make_{i}']

df = df.drop(columns="make")
print(df.info())
```

```python
print("Missing value")
null_data = df[df.isnull().any(axis=1)]
def missing_value():
    if len(null_data) != 0:
        for i in null_data.index:
            l = []
            for j in range(len(df.iloc[i])):
                if np.isnan(df.iloc[i][j]):
                    l.append(df.columns.values[j])
            print("index", i, l)
    else:
        print("There is no empty data")
missing_value()
```

```python
df['price'] = df['price'].replace(np.NaN, df['price'].mean())
df['bore'] = df['bore'].replace(np.NaN, df['bore'].mean())
df['stroke'] = df['stroke'].replace(np.NaN, df['stroke'].mean())
df['hp'] = df['hp'].replace(np.NaN, df['hp'].mean())
df['peak-rpm'] = df['peak-rpm'].replace(np.NaN, df['peak-rpm'].mean())
df.isnull().sum()
```

```python
null_data = df[df.isnull().any(axis=1)]
missing_value()
```

```python
from sklearn import preprocessing
x = df.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled, columns=df.columns)
df.head()
```

```python
for i in df:
    uniq = len(df[i].unique())
    print(i, "=", uniq)
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='green')

plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 42)
y_kmeans = kmeans.fit_predict(df)
df1 = np.array(df)

# size of image
plt.figure(figsize=(15,7))

# visualizing the clusters
sns.scatterplot(df1[y_kmeans == 0, 0], df1[y_kmeans == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(df1[y_kmeans == 1, 0], df1[y_kmeans == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(df1[y_kmeans == 2, 0], df1[y_kmeans == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', 
                label = 'Centroids',s=300,marker=',')

# labeling
# plt.grid(False)
plt.title('Three clusters')
plt.xlabel('Status_type')
plt.ylabel('values')
plt.legend()
plt.show()
```
